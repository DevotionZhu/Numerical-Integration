<h3> Here, we import numpy for mathematical operations and matplotlib for plotting functions. </h3>

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

<h3>We define a Python function for numerical integration using the trapezoidal rule, trapint(f,a,b,N)
    , where f is the integrand function
    , a and b are the lower and upper limits of integration, respectively, and N is the number of disjoint, equal-length, and closed subintervals $ [x_{i},x_{j}] $ of $ [a,b] $.
</h3>

```python
def trapint(f,a,b,N):
    
    h = (b-a)/N
    x = np.linspace(a,b,N+1)
    
    newsum = 0

    for i in range(0,N):

        sum = f(x[i]) + f(x[i+1])    
        newsum = newsum + sum
    
    totarea = (h/2)*newsum 
    
    return totarea
```

<h3>
We also define a Python function for numerical integration using Simpson's rule, simpint(f,a,b,N).
</h3>

```python
def simpint(f,a,b,N):
    
    h = (b-a)/N
    x = np.linspace(a,b,N+1)
    
    newsum = 0

    for i in range(0,N,2):
    
        x_i = (x[i] + x[i+2])/2  
        sum = f(x[i]) + 4*f(x_i) + f(x[i+2])
        newsum = newsum + sum

    totarea = (h/3)*newsum
    
    return totarea
```

<h3>
The Fresnel Integrals are as follows: <br/> <br/> 
$
\begin{equation}
    C(x) = \int_{0}^{x}\cos(t^2)dt \nonumber \\
    S(x) = \int_{0}^{x}\sin(t^2)dt \nonumber
\end{equation}
$
</h3>

```python

f = lambda x: np.sin(x**2)
g = lambda x: np.cos(x**2)
t = np.linspace(-10,10,1000)
S = simpint(f,0,t,1000)
C = simpint(g,0,t,1000)
```

<h3>The Eurler Spiral is defined as the parametric plot: $ (C(x),S(x)) $.</h3>

```python
plt.xlim(-3,3)
plt.plot(t,S,'red',label = 'S(x)')
plt.plot(t,C,'blue', label = 'C(x)')
plt.plot(C,S,'black',label = 'Euler Spiral')
plt.legend(loc='upper left')
plt.show
```
